It can be observed from the information above that the column names of dataset1 and dataset2 are similiar and the number of columns are the same. The inspection  lead the datasets to be integrated by merging the two dataframes and adopt the dataset 1 to dataset 2. 
The schema conflict are: 
1. The difference of column names (letters and blanks)
2. The unit of measurement of the 'salary' column

Since it need to avoid changing the column names, the two datasets will be merged into one by first keeping the different column name and their data in the new dataframe. The column names that are the same in two datasets will be merged directly.

```python
df = df2[['Id','Source Name','Title','location','Contract Type','Contract Time','Company','Category','Salary per month','OpenDate','CloseDate']]\
.merge(df1[['Id','SourceName','Title','Location','ContractType','ContractTime','Company','Category','Salary per annum','OpenDate','CloseDate']], how = "outer")
df
```

The global dataframe will contain null values that are caused by schema conflict(difference of column names), so the next step is to adopt colunm names that from dataset 1. One special column is 'salary per month' and 'salary per annum'. When adopting the column names, it should make the unit measurement be the same as well as the data.

```python
df['location'] = df['location'].fillna(df['Location'])
df['Contract Type'] = df['Contract Type'].fillna(df['ContractType'])
df['Contract Time'] = df['Contract Time'].fillna(df['ContractTime'])
df['Salary per month'] = df['Salary per month'].fillna(df['Salary per annum']//12)
df['Source Name'] = df['Source Name'].fillna(df['SourceName'])
df3 = df.drop(['Location','ContractType','ContractTime','Salary per annum','SourceName'], axis = 1)
df3
```

The global dataframe only keep the column names of dataset 2 as above. The data in the global dataframe still have data conflict. Then it is needed to adopt the data that are from dataset 1 to dataset2.

## Title

```python
#To chnage all the titles to be in same front format(capitalize each words)
df3["Title"] = df3["Title"].str.title()
df3["Title"].head(10)
```

## Company

```python
#df3["Company"] = df3["Company"].str.title()
#df3["Company"].head(10)
```

```python
for i, data in enumerate(df3.Company):
    if "Limited" in str(data):
        df3.iloc[i,6] = df3.iloc[i,6][:-7] + 'Ltd'
    if "LIMITED" in str(data):
        df3.iloc[i,6] = df3.iloc[i,6][:-7] + 'Ltd'
    if "Ltd" in str(data):
        df3.iloc[i,6] = df3.iloc[i,6][:-3] + 'Ltd'
    if "LTD" in str(data):
        df3.iloc[i,6] = df3.iloc[i,6][:-3] + 'Ltd'
```

## Contract Type

```python
#Expand known abbreviations or acronyms
abb = df3.iloc[:,4]
for i,data in enumerate(abb):
    if data == "full_time":
        new_data1 = data.replace("full_time","ft")
        df3.iloc[i,4] = new_data1
    if data == "n/a":
        new_data2 = data.replace("n/a","non-specified")
        df3.iloc[i,4] = new_data2
    if data == "part_time":
        new_data3 = data.replace("part_time","pt")
        df3.iloc[i,4] = new_data3        
    #print(df1.iloc[i,4])
```

## Contract Time

```python
#Expand known abbreviations or acronyms
abb = df3.iloc[:,5]
for i,data in enumerate(abb):
    if data == "contract":
        new_data1 = data.replace("contract","contr.")
        df3.iloc[i,5] = new_data1
    if data == "n/a":
        new_data2 = data.replace("n/a","non-specified")
        df3.iloc[i,5] = new_data2
    if data == "permanent":
        new_data3 = data.replace("permanent","perm.")
        df3.iloc[i,5] = new_data3  
    #print(df1.iloc[i,5])
```

## OpenDate

```python
#To change the datetime in the same format as the format in dataset 2
import datetime
reg = re.compile("T")
t = df['OpenDate']
for i,data in enumerate(t):
    if reg.search(data) != None:
        date = data.split("T")[0][0:4] + "-" + data.split("T")[0][4:6] + "-" + data.split("T")[0][6:8]
        time = data.split("T")[1][0:2] + ":" + data.split("T")[1][2:4] + ":" + data.split("T")[1][4:6]
        new_date = date + " " + time
        df3.iloc[i,-2] = datetime.datetime.strptime(new_date,'%Y-%m-%d %H:%M:%S')
        #print(df3.iloc[i,-2])
    else:
        df3.iloc[i,-2] = datetime.datetime.strptime(data,'%Y-%m-%d %H:%M:%S')
```

## CloseDate

```python
reg = re.compile("T")
t1 = df['CloseDate']
for i,data in enumerate(t1):
    if reg.search(data) != None:
        date = data.split("T")[0][0:4] + "-" + data.split("T")[0][4:6] + "-" + data.split("T")[0][6:8]
        time = data.split("T")[1][0:2] + ":" + data.split("T")[1][2:4] + ":" + data.split("T")[1][4:6]
        new_date = date + " " + time
        df3.iloc[i,-1] = datetime.datetime.strptime(new_date,'%Y-%m-%d %H:%M:%S')
    else:
        df3.iloc[i,-1] = datetime.datetime.strptime(data,'%Y-%m-%d %H:%M:%S')
```

## Global key and duplicated data

This step is to find out the global key of the global dataframe. Since id is unique in the dataframe, it can not be considered as the global key. When try to use the attributes that are combined together as the key to find out the duplicated data, the combination of 'Source Name','Title','location','Contract Type','Contract Time','Company','Category','Salary per month','OpenDate','CloseDate' are to identify the data and detect the duplicated data. So this combined key is the global key.

```python
df3.duplicated(subset=['Source Name','Title','location','Contract Type','Contract Time','Company','Category','Salary per month','OpenDate','CloseDate'], keep='last').sum()
```

```python
a= df3.duplicated(subset=['Source Name','Title','location','Contract Type','Contract Time','Company','Category','Salary per month','OpenDate','CloseDate'], keep='last')
```

```python
#All the duplicated data in a dataframe, only keep the duplicated data once
duplicate_data = pd.DataFrame(df3.loc[a,:])
duplicate_data
```

```python
#drop the duplicated record using the index
drop = list(duplicate_data.index)
df3 = df3.drop(df3.index[drop])
```
